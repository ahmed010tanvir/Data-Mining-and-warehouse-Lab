```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df = pd.read_csv('/content/drive/MyDrive/Data Mining and warehouse/Project-2/mushrooms.csv')
print(df.head())
```

```python
df_encoded = pd.get_dummies(df, columns=df.columns)
df_encoded
```

```python
# Frequent itemset generation

frequent_itemsets = apriori(df_encoded, min_support=0.3, use_colnames=True)
print(frequent_itemsets)
```

```python
for item in frequent_itemsets.iloc[4]:
  print(item)
```

```python
# Association rule generation

rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)
rules
```

```python
rules = rules.sort_values(by=['confidence', 'lift', 'support'], ascending=False)
rules
```
